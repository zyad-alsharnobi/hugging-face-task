```python
import requests
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def query(filename):
    API_URL = "https://api-inference.huggingface.co/models/facebook/detr-resnet-50"
    headers = {"Authorization": "Bearer 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"}
    with open(filename, "rb") as f:
        data = f.read()
    response = requests.post(API_URL, headers=headers, data=data)
    return response.json()

def visualize_detections(image_path, detections):
    # Load and display the image
    img = Image.open(image_path)
    fig, ax = plt.subplots(1)
    ax.imshow(img)
    
    # Define colors for different confidence levels
    colors = ['r', 'g', 'b', 'y', 'c', 'm']
    
    # Draw bounding boxes and labels
    for idx, detection in enumerate(detections):
        box = detection['box']
        label = detection['label']
        score = detection['score']
        
        # Create rectangle patch
        rect = patches.Rectangle(
            (box['xmin'], box['ymin']),
            box['xmax'] - box['xmin'],
            box['ymax'] - box['ymin'],
            linewidth=2,
            edgecolor=colors[idx % len(colors)],
            facecolor='none'
        )
        
        # Add rectangle to plot
        ax.add_patch(rect)
        
        # Add label with confidence score
        plt.text(
            box['xmin'], box['ymin'] - 5,
            f'{label}: {score:.2f}',
            color=colors[idx % len(colors)],
            fontsize=10,
            bbox=dict(facecolor='white', alpha=0.8)
        )
    
    plt.axis('off')
    plt.show()

# Run detection and visualize results
image_path = "generated_image.jpg"
detections = query(image_path)
visualize_detections(image_path, detections)
```
